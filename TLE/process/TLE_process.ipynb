```python
# import packages

import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import pandas as pd
```

```python
# import data

df = pd.read_csv("data.txt", sep = " ")
df
```

```python
fig = go.Figure(go.Scatter3d(
    z = df['ECI_x'],
    y = df['ECI_y'],
    x = df['ECI_z'],
    
    marker = dict(
        #color = 'rgb(128, 128, 128)',
        size = 2,
        symbol = 'circle'
    )
))

fig.update_layout(
    scene = {
        "xaxis": {"nticks": 20, "range": [-8000, 8000]},
        "yaxis": {"nticks": 4, "range": [-8000, 8000]},
        "camera_eye": {"x": 1, "y": 0.8, "z": 0.8},
        "aspectratio": {"x": 1, "y": 1, "z": 0.1}
    }
)

fig.show()
```
